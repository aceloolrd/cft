```python
def get_unique_data_types(folder_path: str) -> set:
    """
    Получает уникальные типы данных из всех файлов CSV в указанной папке.

    Args:
        - folder_path (str): Путь к папке с файлами CSV.

    Returns:
        - set: Множество уникальных типов данных.
    """
    unique_data_types = set()
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.csv'):
            file_path = os.path.join(folder_path, file_name)
            df = pd.read_csv(file_path)
            unique_data_types.update(df.dtypes)
    return unique_data_types

```

Далее представлены функции, с которыми я работал в рамках решения данной задачи.

```python
def get_max_string_lengths_object_types(df: pd.DataFrame) -> dict:
    """
    Находит максимальные длины записей в object признаках в DataFrame.

    Args:
        - df (pd.DataFrame): DataFrame для анализа.

    Returns:
        - dict: Словарь, содержащий максимальные длины строковых типов данных.
    """
    type_varchar = {}
    for column in df.select_dtypes(include='object').columns:
        arr = df[column].unique()
        max_length = max(len(str(element)) for element in arr)
        type_varchar[column] = max_length + 5
        
    return type_varchar


def get_integer_types(df: pd.DataFrame) -> dict:
    """
    Определяет типы данных для БД для целочисленных столбцов из DataFrame.

    Args:
        - df (pd.DataFrame): DataFrame для анализа.

    Returns:
        - dict: Словарь, содержащий типы целочисленных столбцов.
    """
    type_int = [column for column in df.columns if df[column].dtype == 'int64']
    type_integer = {}
    for column in type_int:
        maximum = df[column].max()
        minimum = df[column].min()
        if maximum < 2**15 and minimum > -(2**15):
            type_integer[column] = 'smallint'
        elif maximum < 2**31 and minimum > -(2**31):
            type_integer[column] = 'integer'
        elif maximum < 2**63 and minimum > -(2**63):
            type_integer[column] = 'bigint'
        else:
            type_integer[column] = 'numeric'  # если значения выходят за диапазон целочисленных типов
            
    return type_integer


def get_float_types(df: pd.DataFrame) -> dict:
    """
    Определяет типы данных для БД для столбцов с плавающей точкой в DataFrame.

    Args:
        - df (pd.DataFrame): DataFrame для анализа.

    Returns:
        - dict: Словарь, содержащий типы столбцов с плавающей точкой.
    """
    type_float = [column for column in df.columns if df[column].dtype == 'float64']
    type_real = {}
    for column in type_float:
        length = df[column].map(lambda x: len(str(x))).max() # быстрее, чем length = df[column].astype(str).map(len).max()
        if length < 38:
            type_real[column] = 'real'
        else:
            type_real[column] = 'float'
            
    return type_real


def analyze_dataframe(df: pd.DataFrame) -> tuple[dict, dict, dict]:
    """
    Вызывает функции для определения типов данных для каждого столбца в будущей БД.

    Args:
        - df (pd.DataFrame): DataFrame для анализа.

    Returns:
        - tuple[dict, dict, dict]: Кортеж, содержащий словари с типами данных для строковых, целочисленных и числовых столбцов.
    """
    # Вызов функций анализа для строковых, целочисленных и числовых типов данных
    object_types = get_max_string_lengths_object_types(df)
    integer_types = get_integer_types(df)
    float_types = get_float_types(df)
    
    return object_types, integer_types, float_types


def generate_sql_schema(table_name: str, column_order: list, object_types: dict, integer_types: dict, float_types: dict) -> str:
    """
    Генерирует запрос SQL для создания схемы таблицы на основе DataFrame.

    Args:
        - table_name (str): Название таблицы.
        - column_order (list): Порядок столбцов в таблице.
        - object_types (dict): Типы строковых столбцов.
        - integer_types (dict): Типы целочисленных столбцов.
        - float_types (dict): Типы столбцов с плавающей точкой.

    Returns:
        - str: Запрос SQL для создания схемы таблицы.
    """
    sql_schema_query = f"DROP TABLE IF EXISTS {table_name};\nCREATE TABLE {table_name}(\n"
    
    # Добавление определений столбцов в порядке, заданном column_order
    for column in column_order:
        if column in object_types:
            length = object_types[column]
            sql_schema_query += f"  {column} varchar({length}),\n"
        elif column in integer_types:
            data_type = integer_types[column]
            sql_schema_query += f"  {column} {data_type},\n"
        elif column in float_types:
            data_type = float_types[column]
            sql_schema_query += f"  {column} {data_type},\n"
    
    # Удаление последней запятой и добавление закрывающей скобки
    sql_schema_query = sql_schema_query.rstrip(',\n') + '\n);'
    
    return sql_schema_query


def find_primary_keys(df: pd.DataFrame) -> list:
    """
    Поиск столбцов с уникальными значениями.

    Args:
        - df (pd.DataFrame): DataFrame для анализа.

    Returns:
        - list: Список возможных первичных ключей.
    """
    unique_columns = []
    for column in df.columns:
        if df[column].is_unique:
            unique_columns.append(column)
    return unique_columns


def generate_sql_from_dataframe(df: pd.DataFrame, table_name: str) -> tuple[str, str]:
    """
    Генерирует запросы SQL для создания таблицы и загрузки данных из DataFrame.

    Args:
        - df (pd.DataFrame): DataFrame для анализа.
        - table_name (str): Название таблицы.

    Returns:
        - tuple[str, str]: Кортеж с запросами SQL для создания таблицы и загрузки данных.
    """
    column_order = list(df.columns)
    
    sql_schema_query = generate_sql_schema(table_name, column_order, *analyze_dataframe(df))
    sql_data_query = f"COPY {table_name}\n  ({',\n  '.join(column_order)})\nFROM '{DATA_FULL_PATH}{table_name}.csv' DELIMITER ',' CSV HEADER;"
    
    return sql_schema_query, sql_data_query


def generate_sql_keys_query(table_name: str, primary_key: str = None, foreign_keys: list = None) -> str:
    """
    Генерирует запросы SQL для определения ключей таблицы.

    Args:
        - table_name (str): Название таблицы.
        - primary_key (str): Первичный ключ.
        - foreign_keys (list): Список внешних ключей.

    Returns:
        - str: Запрос SQL для определения ключей таблицы.
    """
    sql_query = ""
    
    # Добавление определения первичного ключа, если указан
    if primary_key:
        sql_query += f"ALTER TABLE {table_name}\nADD PRIMARY KEY ({primary_key});\n"
    
    # Добавление определений внешних ключей, если указаны
    if foreign_keys:
        for column, reference_table, reference_column in foreign_keys:
            constraint_name = f"fk_{table_name}_{column}"
            sql_query += f"ALTER TABLE {table_name}\nADD CONSTRAINT {constraint_name}\n"
            sql_query += f"FOREIGN KEY ({column})\nREFERENCES {reference_table}({reference_column});\n"
    
    return sql_query

# не стал ничего удалять
def remove_missing_values(df_main: pd.DataFrame, df_reference: pd.DataFrame, column_name: str) -> pd.DataFrame:
    """
    Удаляет строки из основного DataFrame, которые имеют значения в столбце, отсутствующие в ссылочном DataFrame.

    Args:
        - df_main (pd.DataFrame): Основной DataFrame, из которого нужно удалить строки.
        - df_reference (pd.DataFrame): Ссылочный DataFrame, значения которого используются для проверки.
        - column_name (str): Имя столбца, по значениям которого осуществляется проверка.

    Returns:
        - pd.DataFrame: DataFrame с удаленными строками.
    """
    column_values_main = df_main[column_name]
    column_values_reference = df_reference[column_name]
    values_to_delete = column_values_main[~column_values_main.isin(column_values_reference)].tolist()
    return df_main[~df_main[column_name].isin(values_to_delete)]

```

```python
#7 вывести тех клиентов, у кого доход на одного члена семьи в два раза больше, чем в среднем на одного члена семьи по выборке
query = """
SELECT
    sk_id_curr AS client_id
FROM 
    application AS a
WHERE
    (a.AMT_INCOME_TOTAL / a.CNT_FAM_MEMBERS) / 2 > (SELECT AVG(b.AMT_INCOME_TOTAL / b.CNT_FAM_MEMBERS) FROM application AS b)
"""

db_connector.get_df_from_query(query)
```

```python
#8 вывести клиентов старше 60 лет по которым нет данных в bureau
query = """
SELECT 
    application.sk_id_curr AS clients
FROM application
LEFT JOIN
    bureau
ON
    application.sk_id_curr = bureau.sk_id_curr
WHERE
    bureau.sk_id_curr IS NULL AND
    days_birth < -21915
"""

db_connector.get_df_from_query(query)
```

```python
#9 вывести женщин, у которых в истории bureau было больше двух кредитов, просроченных на 61 день и более
# отсортировать в порядке убывания по кол-ву таких кредитов
query = """
WITH bureau_table AS
    (SELECT sk_id_curr, SUM(status_overdue) AS overdue_credit
     FROM bureau
     JOIN
        (SELECT sk_id_bureau, 1 AS status_overdue
         FROM bureau_balance
         WHERE status in ('3','4','5')
         GROUP BY sk_id_bureau) AS bureau_balance_over
         ON bureau.sk_id_bureau = bureau_balance_over.sk_id_bureau
     GROUP BY sk_id_curr)

SELECT application.sk_id_curr AS women, overdue_credit
FROM application
JOIN bureau_table
ON
application.sk_id_curr = bureau_table.sk_id_curr
WHERE
code_gender = 'F' AND overdue_credit > 2
ORDER BY overdue_credit DESC
"""

db_connector.get_df_from_query(query)
```

```python
#10 по данным из bureau (БКИ) расчитать долю просрочки в активных займах для каждого клиента
# вывести топ 7 мужчин с наибольшей суммой просрочки, указав для них, помимо прочего,
# сумму активных кредитов и суммы всех кредитов (активных и закрытых).
query = """
WITH overdue_info AS (
    SELECT 
        b.sk_id_curr,
        SUM(CASE 
                WHEN b.credit_active = 'Active' 
                THEN b.amt_credit_sum_overdue 
                ELSE 0 
            END) AS overdue_sum,
        SUM(CASE 
                WHEN b.credit_active = 'Active' 
                THEN b.amt_credit_sum 
                ELSE 0 
            END) AS active_sum,
        CASE 
            WHEN SUM(CASE 
                        WHEN b.credit_active = 'Active' 
                        THEN b.amt_credit_sum 
                        ELSE 0 
                    END) = 0 
            THEN 0 
            ELSE COALESCE(SUM(CASE 
                                WHEN b.credit_active = 'Active' 
                                THEN b.amt_credit_sum_overdue 
                                ELSE 0 
                            END) / SUM(CASE 
                                            WHEN b.credit_active = 'Active' 
                                            THEN b.amt_credit_sum 
                                            ELSE 0 
                                        END), 0)
        END AS overdue_ratio,
        SUM(b.amt_credit_sum) AS total_sum
    FROM 
        bureau b
    GROUP BY 
        b.sk_id_curr
)
SELECT 
    o.sk_id_curr,
    o.overdue_ratio,
    o.overdue_sum,
    o.active_sum,
    o.total_sum
FROM 
    overdue_info o
JOIN 
    application a ON a.sk_id_curr = o.sk_id_curr
WHERE 
    a.code_gender = 'M'
ORDER BY 
    o.overdue_sum DESC
LIMIT 7;
"""

db_connector.get_df_from_query(query)
```

```python
#2.  Женщины без детей с машиной и недвижимостью 1 иначе 0
query = """
SELECT
    sk_id_curr AS client,
    CASE WHEN (flag_own_car='Y' AND code_gender = 'F' AND flag_own_realty = 'Y' AND cnt_children = 0)  
         THEN 1 
         ELSE 0 
         END AS women_with_assets
FROM application

"""
db_connector.get_df_from_query(query)
```

```python
#3.  Возраст заемщика 
sql_query = """
    SELECT DISTINCT sk_id_curr, (days_birth*(-1))/365 AS age
    FROM application
    ORDER BY sk_id_curr;
"""
db_connector.get_df_from_query(sql_query)
```

```python
#4.  Наличие/отсутствие высшего образования
sql_query = """
    SELECT DISTINCT sk_id_curr, 
    	   (CASE 
    	   	 WHEN name_education_type = 'Academic degree' OR name_education_type = 'Higher education'
    	   	 THEN 1
    	   	 ELSE 1
    	     END) AS flag_higher_education
    FROM application
    ORDER BY sk_id_curr;
"""
db_connector.get_df_from_query(sql_query)
```

```python
#5. Отношение дохода на 1 члена семьи к выплате по кредиту
query = """
SELECT 
    sk_id_curr AS client,
    (amt_income_total / cnt_fam_members) / amt_annuity AS income_annuity_ratio
FROM application
"""

db_connector.get_df_from_query(query)
```
